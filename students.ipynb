# Loading the Dataset: Students.csv

```python
students_df = pd.read_csv("Students.csv", delimiter=';')
```

# Inspecting the Dataset:

```python
students_df.head()
```

# Question (a): Probability of a Student Being a "Graduate":

```python
graduate_probability = students_df['Target'].value_counts(normalize=True).get('Graduate', 1)
graduate_probability
```

# Question (b): Probability of a Student Being a "Dropout" with "Marital status" of 1:

```python
dropout_with_marital_status_1 = students_df[(students_df['Target'] == 'Dropout') & (students_df['Marital status'] == 5)]
dropout_probability = len(dropout_with_marital_status_1) / len(students_df)
dropout_probability
```

# Question (c): Relationship between "Target" and "Mother's Qualification":

```python
relationship = pd.crosstab(students_df['Mother\'s qualification'], students_df['Target'], normalize='columns')
relationship
```
